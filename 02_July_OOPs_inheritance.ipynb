```python
1.Explain what inheritance is in object-oriented programming and why it is used?
ans:In object-oriented programming (OOP), inheritance is a mechanism that allows one class (called the child or derived class) to inherit properties and behaviors from another class (called the parent or base class). This means that the child class can reuse code defined in the parent class without having to re-implement it.
    Parent Class (Base Class): This is the class that provides the initial set of properties and methods that are inherited by other classes. It serves as a blueprint.
    Child Class (Derived Class): This class inherits attributes and methods from its parent class. It can also extend or modify these attributes and methods, and add its own unique attributes and methods.
    We can use inheritance when one class is a specialized version of another (e.g., Dog is a type of Animal)and Used to avoid duplicating code across multiple classes that share common attributes and behaviors.    
```

```python
5.What is the purpose of the "super" keyword in inheritance? Provide an example.
ans:Super keyword allows a subclass to access methods, constructors, and properties defined in its superclass and Used to invoke the constructor of the superclass from the constructor of the subclass, especially when the subclass needs to initialize inherited fields.
    class Animal {
    String species;

    Animal(String species) {
        this.species = species;
    }

    void eat() {
        System.out.println("Animal is eating");
    }
}

class Dog extends Animal {
    String breed;

    Dog(String species, String breed) {
        super(species);  // Calling superclass constructor
        this.breed = breed;
    }

    void bark() {
        System.out.println("Dog is barking");
    }

    void displayDetails() {
        System.out.println("Species: " + species);
        System.out.println("Breed: " + breed);
    }
}

public class Main {
    public static void main(String[] args) {
        Dog myDog = new Dog("Canine", "Labrador");
        myDog.eat();  // Accessing inherited method from superclass
        myDog.bark();  // Accessing method from subclass
        myDog.displayDetails();  // Accessing method from subclass
    }
}
Animal is the superclass with a constructor that initializes the species attribute.Dog is a subclass of Animal that extends its functionality and introduces a new attribute breed.In the constructor of Dog, super(species) calls the constructor of Animal, ensuring that the species attribute is initialized properly.
The super keyword also allows Dog to invoke the eat() method defined in Animal directly (super.eat() could be used within Dog if eat() were marked as protected or public in the Animal class).
```

```python
6.Create a base class called "Vehicle" with attributes like "make", "model", and "year".Then, create a derived class called "Car" that inherits from "Vehicle" and adds an
attribute called "fuel_type". Implement appropriate methods in both classes.
ans:// Base class
class Vehicle {
    // Attributes
    private String make;
    private String model;
    private int year;

    // Constructor
    public Vehicle(String make, String model, int year) {
        this.make = make;
        this.model = model;
        this.year = year;
    }

    // Getters and setters
    public String getMake() {
        return make;
    }

    public void setMake(String make) {
        this.make = make;
    }

    public String getModel() {
        return model;
    }

    public void setModel(String model) {
        this.model = model;
    }

    public int getYear() {
        return year;
    }

    public void setYear(int year) {
        this.year = year;
    }

    // Method to display vehicle information
    public void displayInfo() {
        System.out.println("Make: " + make);
        System.out.println("Model: " + model);
        System.out.println("Year: " + year);
    }
}

// Derived class
class Car extends Vehicle {
    // Additional attribute
    private String fuelType;

    // Constructor
    public Car(String make, String model, int year, String fuelType) {
        super(make, model, year); // Call to superclass constructor
        this.fuelType = fuelType;
    }

    // Getter and setter for fuelType
    public String getFuelType() {
        return fuelType;
    }

    public void setFuelType(String fuelType) {
        this.fuelType = fuelType;
    }

    // Method to display car information, overriding displayInfo() method in Vehicle
    @Override
    public void displayInfo() {
        super.displayInfo();  // Call to superclass method
        System.out.println("Fuel Type: " + fuelType);
    }
}

// Main class to test Vehicle and Car classes
public class Main {
    public static void main(String[] args) {
        // Creating a Vehicle object
        Vehicle vehicle = new Vehicle("Honda", "Civic", 2022);
        vehicle.displayInfo();
        System.out.println();

        // Creating a Car object
        Car car = new Car("Toyota", "Corolla", 2020, "Gasoline");
        car.displayInfo();
    }
}
```

```python
8.Design a base class called "Shape" with attributes like "colour" and "border_width."Create derived classes, "Rectangle" and "Circle," that inherit from "Shape" and add
specific attributes like "length" and "width" for the "Rectangle" class and "radius" for the "Circle" class.
ans:// Base class
class Shape {
    // Attributes
    private String colour;
    private int borderWidth;

    // Constructor
    public Shape(String colour, int borderWidth) {
        this.colour = colour;
        this.borderWidth = borderWidth;
    }

    // Getters and setters
    public String getColour() {
        return colour;
    }

    public void setColour(String colour) {
        this.colour = colour;
    }

    public int getBorderWidth() {
        return borderWidth;
    }

    public void setBorderWidth(int borderWidth) {
        this.borderWidth = borderWidth;
    }

    // Method to display shape information
    public void displayInfo() {
        System.out.println("Colour: " + colour);
        System.out.println("Border Width: " + borderWidth);
    }
}

// Derived class Rectangle
class Rectangle extends Shape {
    // Additional attributes
    private double length;
    private double width;

    // Constructor
    public Rectangle(String colour, int borderWidth, double length, double width) {
        super(colour, borderWidth); // Call to superclass constructor
        this.length = length;
        this.width = width;
    }

    // Getters and setters for Rectangle
    public double getLength() {
        return length;
    }

    public void setLength(double length) {
        this.length = length;
    }

    public double getWidth() {
        return width;
    }

    public void setWidth(double width) {
        this.width = width;
    }

    // Method to calculate area of Rectangle
    public double calculateArea() {
        return length * width;
    }

    // Method to display rectangle information, overriding displayInfo() method in Shape
    @Override
    public void displayInfo() {
        super.displayInfo();  // Call to superclass method
        System.out.println("Length: " + length);
        System.out.println("Width: " + width);
        System.out.println("Area: " + calculateArea());
    }
}

// Derived class Circle
class Circle extends Shape {
    // Additional attribute
    private double radius;

    // Constructor
    public Circle(String colour, int borderWidth, double radius) {
        super(colour, borderWidth); // Call to superclass constructor
        this.radius = radius;
    }

    // Getter and setter for Circle
    public double getRadius() {
        return radius;
    }

    public void setRadius(double radius) {
        this.radius = radius;
    }

    // Method to calculate area of Circle
    public double calculateArea() {
        return Math.PI * radius * radius;
    }

    // Method to display circle information, overriding displayInfo() method in Shape
    @Override
    public void displayInfo() {
        super.displayInfo();  // Call to superclass method
        System.out.println("Radius: " + radius);
        System.out.println("Area: " + calculateArea());
    }
}

// Main class to test Shape, Rectangle, and Circle classes
public class Main {
    public static void main(String[] args) {
        // Creating a Rectangle object
        Rectangle rectangle = new Rectangle("Red", 2, 5.0, 3.0);
        rectangle.displayInfo();
        System.out.println();

        // Creating a Circle object
        Circle circle = new Circle("Blue", 1, 4.0);
        circle.displayInfo();
    }
}
```

```python
10.Create a base class called "BankAccount" with attributes like "account_number" and "balance." Derive two classes, "SavingsAccount" and "CheckingAccount," from
"BankAccount." Add specific methods like "calculate_interest" for the "SavingsAccount" class and "deduct_fees" for the "CheckingAccount" class.
ans:class BankAccount:
    def __init__(self, account_number, balance=0):
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited ${amount}. New balance is ${self.balance}.")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew ${amount}. New balance is ${self.balance}.")
        else:
            print("Insufficient funds or invalid withdrawal amount.")

    def get_balance(self):
        return self.balance


class SavingsAccount(BankAccount):
    def __init__(self, account_number, balance=0, interest_rate=0.02):
        super().__init__(account_number, balance)
        self.interest_rate = interest_rate

    def calculate_interest(self):
        interest = self.balance * self.interest_rate
        print(f"Calculated interest: ${interest}.")
        return interest


class CheckingAccount(BankAccount):
    def __init__(self, account_number, balance=0, fee=5):
        super().__init__(account_number, balance)
        self.fee = fee

    def deduct_fees(self):
        if self.balance >= self.fee:
            self.balance -= self.fee
            print(f"Deducted fees: ${self.fee}. New balance is ${self.balance}.")
        else:
            print("Insufficient funds to deduct fees.")


# Example usage
savings = SavingsAccount("SA12345", 1000)
savings.deposit(200)
savings.withdraw(100)
interest = savings.calculate_interest()
print(f"Interest: ${interest}")

checking = CheckingAccount("CA12345", 500)
checking.deposit(150)
checking.withdraw(100)
checking.deduct_fees()
```
